```python
import numpy as np
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
cols = ["Sex","Length","Diameter","Height","Whole_weight",
        "Shucked_weight","Viscera_weight","Shell_weight","Rings"]

df = pd.read_csv(url, header=None, names=cols)

print("\nDATASET DETAILS")
print("Total rows :", len(df))
print("\nColumns :")
for c in df.columns:
    print(c)
print("\nFirst 5 rows :")
print(df.head())

y = df["Rings"].values.reshape(-1,1) + 1.5
X = df[["Length","Diameter","Shell_weight"]].values

N = len(X)
cut = int(0.8 * N)

X_train = X[:cut]
X_test = X[cut:]
y_train = y[:cut]
y_test = y[cut:]

print("\nTRAIN TEST SPLIT")
print("X_train shape :", X_train.shape)
print("X_test shape  :", X_test.shape)
print("y_train shape :", y_train.shape)
print("y_test shape  :", y_test.shape)

m = X_train.mean(axis=0)
s = X_train.std(axis=0)

X_train = (X_train - m) / s
X_test = (X_test - m) / s

def forward(X, w, b):
    return np.dot(X, w) + b

def mse(y, y_hat):
    return np.mean((y - y_hat) ** 2)

def mae(y, y_hat):
    return np.mean(np.abs(y - y_hat))

def grad_w(X, y, y_hat):
    n = len(X)
    return (-2/n) * np.dot(X.T, (y - y_hat))

def grad_b(y, y_hat):
    n = len(y)
    return (-2/n) * np.sum(y - y_hat)

d = X_train.shape[1]
w = np.random.randn(d,1) * 0.01
b = 0.0

lr = 0.01
epochs = 200

print("\nTraining started\n")

for e in range(epochs):
    y_hat = forward(X_train, w, b)
    loss = mse(y_train, y_hat)
    dw = grad_w(X_train, y_train, y_hat)
    db = grad_b(y_train, y_hat)
    w = w - lr * dw
    b = b - lr * db
    if e % 20 == 0:
        print("Epoch", e, "Loss :", round(loss,6))

print("\nTraining completed")

y_test_hat = forward(X_test, w, b)

test_mse = mse(y_test, y_test_hat)
test_mae = mae(y_test, y_test_hat)

print("\nEvaluation on test data")
print("Test MSE :", round(test_mse,6))
print("Test MAE :", round(test_mae,6))

print("\nSample predictions\n")

for i in range(5):
    t = y_test[i][0]
    p = y_test_hat[i][0]
    err = abs(t - p)
    print("Sample", i+1)
    print("True age :", round(t,3))
    print("Predicted age :", round(p,3))
    print("Absolute error :", round(err,3))
    print()
```
