```python
import numpy as np

np.random.seed(42)


X = np.linspace(0, 10, 50).reshape(-1, 1)
noise = np.random.normal(0, 0.2, size=(50, 1))
y = np.log(X + 1) + noise

print(f"X shape: {X.shape}, y shape: {y.shape}")


hidden_units = 3
print(f"Hidden Units: {hidden_units}")


W1 = np.random.uniform(-1, 1, size=(1, hidden_units))
b1 = np.zeros((1, hidden_units))

W2 = np.random.uniform(-1, 1, size=(hidden_units, 1))
b2 = np.zeros((1, 1))

print(f"W1 shape: {W1.shape}, b1 shape: {b1.shape}")
print(f"W2 shape: {W2.shape}, b2 shape: {b2.shape}")

def activation(z):
    return np.maximum(0, z)

def activation_slope(z):
    return (z > 0).astype(float)


learning_rate = 0.01
epochs = 1000

print("\nStarting training...")
for epoch in range(epochs):

    # Forward pass
    z1 = X @ W1 + b1
    h = activation(z1)
    y_hat = h @ W2 + b2

    # loss
    error = y_hat - y
    loss = np.mean(error ** 2)

    # Backward pass (gradients)
    dL_dy = 2 * error / len(X)
    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh = dL_dy @ W2.T
    dL_dz1 = dL_dh * activation_slope(z1)

    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

    # Update parameters
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2

    if epoch % 100 == 0:
        print(f"epoch {epoch:04d} | loss: {loss:.6f}")


print(f"final Loss: {loss:.6f}")


print("\n Sample :")
for i in range(5):
    print(f"X: {X[i][0]:.2f} | True: {y[i][0]:.4f} | Pred: {y_hat[i][0]:.4f}")
```
